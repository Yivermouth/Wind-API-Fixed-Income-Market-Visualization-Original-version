```python
from WindPy import w
import pandas as pd
import numpy as np
import datetime 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
#以下为解决中文无法显示的问题
#针对matplotlib
plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
w.start()
```

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~设置起、止日期~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

enddate = datetime.date.today()

startdate = enddate + datetime.timedelta(days=-7)       #“-7”代表往前7天

start_date = startdate.strftime('%Y/%m/%d')    #加不加“%H:%M:%S”无所谓

end_date = enddate.strftime('%Y/%m/%d')

print(start_date,type(start_date),end_date,type(end_date))
```

```python
#【注意】——这里取数范围为【All Bonds】

all_bond_data = w.wset("newbondissueview",
       "startdate="+start_date+";enddate="+end_date+";datetype=startdate;bondtype=all;dealmarket=allmarkets;maingrade=all",
      usedf=True)
all_bond_data = all_bond_data[1]
all_bond_data.tail(7)
```

```python
all_bond_data["Wind债券一级分类"]=w.wss(list(all_bond_data["windcode"].values), "windl1type",usedf=True)[1].values
all_bond_data["Wind债券二级分类"]=w.wss(list(all_bond_data["windcode"].values), "windl2type",usedf=True)[1].values
all_bond_data
# all_bond_data.to_excel("本周发行所有债券.xlsx")
```

## 以上为【全部Bond】，现仅筛选【信用债】

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~由于all_bond_data是【全部bond】，使用以下2种方法筛选出【信用债】~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

# 如果你要筛选的类别比较多，可以把它们放在一个列表中，然后使用 isin 方法来进行筛选——chatgpt
categories = ["短期融资券", "企业债", "公司债", "中期票据"]

#【方法一】——使用df[df['Wind债券一级分类'].isin(categories)]
# credit_bond_data[credit_bond_data['Wind债券一级分类'].isin(categories)]

#【方法二】——使用df.query
credit_bond_data = all_bond_data.query('`Wind债券一级分类`.isin(@categories)') 

#~~~~~~~~~~~~由于使用df.query筛选后index不会自动调整。所以要重新设置index~~~~~~~~~~~~~~~~
credit_bond_data = credit_bond_data.reset_index(drop=True)#——reset_index 方法默认保留原索引，可以通过设置drop=True删除原索引

credit_bond_data.index = credit_bond_data.index + 1 #index不设置默认从“0”开始
# credit_bond_data.index.name = "Index"

credit_bond_data
# credit_bond_data.to_excel("本周信用债（含未发行或未及时披露信息）.xlsx")
```

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~添加【其他信息】~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

credit_bond_data["发行量（亿）"] = w.wss(list(credit_bond_data["windcode"].values), "issueamount", usedf=True)[1].values
credit_bond_data["发行量（亿）"] = credit_bond_data["发行量（亿）"]/100000000

credit_bond_data["主体评级"] = w.wss(list(credit_bond_data["windcode"].values), "issuer_rating", usedf=True)[1].values
credit_bond_data["票面利率"] = w.wss(list(credit_bond_data["windcode"].values), "couponrate", usedf=True)[1].values
credit_bond_data["发行人性质"] = w.wss(list(credit_bond_data["windcode"].values), "abs_industry1", usedf=True)[1].values
credit_bond_data["发行人Wind行业(一级)"] = w.wss(list(credit_bond_data["windcode"].values), "abs_industry", usedf=True)[1].values
credit_bond_data["发行人省份"] = w.wss(list(credit_bond_data["windcode"].values), "province", usedf=True)[1].values

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~☆☆☆☆☆【来自Wind官方客服】~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
#【债券代码为：“d”或者“q”开头的】——发行金额、票面利率取值为NaN

# 这些是还没有发行上市的，d或者q开头是临时代码，还没有正式债券代码，一般发行成功之后，交易所等披露正式债券代码之后，可提取相关数据 

# 这些d或者q开头的债券，统计本周发行债券时不要纳入

# 还有一些即使不是“d”或者“q”开头的，但仍然取值为Nan，一般为还无相关信息披露（各种原因都有，具体原因无法知晓）
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

# credit_bond_data = credit_bond_data.dropna() 

credit_bond_data
# credit_bond_data.to_excel("信用债发行data（剔除Nan值）.xlsx")
```
